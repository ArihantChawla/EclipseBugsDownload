```python
for i in range(2):
    url = str(int(df['Bug ID'][i]))
    url = base_url + url
    time.sleep(2)
    user_agent = random.choice(user_agent_list)
    page = requests.get(url,headers = {'User-Agent': user_agent})
    soup = bs(page.text,'lxml')
    hmm = soup.select("tr")
    hmm = hmm[1:]
    text = ""
    count = 0
    for k in hmm:
        #print(k)
        if count == 0:
            k = str(k)
            j = k.split('</td>')
            t = j[2]
            if "Status" in t:
                count = 1
                m = j[3]
                m = m.split('<td>')
                m = m[1]
                m = m.split('\n')
                m = m[0]
                text = text + str(m) + "-->"
                m = j[4]
                m = m.split('<td>')
                m = m[1]
                m = m.split('\n')
                m = m[0]
                text = text + str(m) + "-->"        
        else:
            k = str(k)
            j = k.split('</td>')
            t = j[2]
            if "Status" in t:
                m = j[4]
                m = m.split('<td>')
                m = m[1]
                m = m.split('\n')
                m = m[0]
                text = text + str(m) + "-->"
    text = text[:len(text)-3]
    print(i,text)
    df.at[i,'Activity'] = text
```
