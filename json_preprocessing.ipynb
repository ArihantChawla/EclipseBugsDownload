```python
for q in os.listdir("./folders"):
    path = './folders/' + q + "/"
    for u in os.listdir(path):
        if u != 'urlslvl2.txt':
            ph = path + u + "/"
            for v in os.listdir(ph):
                if v.endswith('.json'):
                    if not v.endswith('_mod.json'):
                        p = ph + v
                        f = open(p,'r')
                        a = f.read()
                        print(p)
                        f.close()
                        j = a.split('"bug":')
                        j = j[1]
                        k = j.split(', "cc":')
                        k = k[0]
                        k = k.split('"comment_sort_order":')
                        k = k[0]
                        k = k + '}'
                        l = p.split('.json')
                        l = l[0]
                        l = l + "_mod.json"
                        f = open(l,'w')
                        f.write(k)
                        f.close()
```
