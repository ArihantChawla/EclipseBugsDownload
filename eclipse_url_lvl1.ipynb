```python
from bs4 import BeautifulSoup as bs
import lxml
import requests 
import os
```

```python
page = requests.get("https://bugs.eclipse.org/bugs/describecomponents.cgi")
```

```python
soup = bs(page.text,'lxml')
```

```python
soup.prettify
```

```python
tr = soup.select("th > a")
tr = str(tr)
print(tr)
```

```python
tr =  tr.split(",")
url = "https://bugs.eclipse.org/bugs/"
```

```python
with open("/home/arihant/Desktop/BUGZZZZ/eclipse/urlslvl1.txt", "a") as myfile:
    for i in tr:
        i = i.split('"')
        i = url + i[1] + '\n'
        myfile.write(i)
```
