```python
for q in os.listdir("./folders"):
    path = "folders/" + q + "/urlslvl2.txt"
    print(path)
    with open(path) as f:
        urls = f.readlines()
        for i in urls:
            component = i.split("=")
            component = component[2]
            component = component.split("&")
            component = component[0]
            ph = "folders/" + q + "/" + component + "/"
            file = ph + "/urlslvl3.txt"
            time.sleep(2)
            user_agent = random.choice(user_agent_list)
            page = requests.get(i,headers = {'User-Agent': user_agent})
            soup = bs(page.text,'lxml')
            t = soup.select("tr > td > a")
            t = str(t)
            t = t.split(",")
        #   print(t)
            u = []
            for k in t:
                try:
                    k = k.split('"')
                    k = k[1] 
        #        k = k.split(">")
        #        k = k[0]
                    k = str(k)
                    if k[0] == 's':
                        u.append(str(k))
                except:
                    pass
            for k in u:
                with open(file, "a") as myfile:
                    k = base_url + k + "\n"
                    myfile.write(k)   
```

```python
for q in os.listdir("./folders"):
    path = "folders/" + q + "/"
    for w in os.listdir(path):
        if w != 'urlslvl2.txt':
            w = path + w + "/"
            infilename = w + "urlslvl3.txt"
            outfilename = w + "urlslvl3_modified.txt"
            lines_seen = set() # holds lines already seen
            outfile = open(outfilename, "w")
            try:
                for line in open(infilename, "r"):
                    tukde = line.split('?')
                    line = tukde[0] + "?format=multiple&" + tukde[1]
                    if line not in lines_seen: # not a duplicate
                        outfile.write(line)
                        lines_seen.add(line)
            except:
                pass
            outfile.close()
#            print(file)
```
