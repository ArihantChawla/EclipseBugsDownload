```python
import pandas as pd
import datetime
import numpy as np
```

```python
df = pd.read_csv('final.csv')
```

```python
df
```

```python
for i in range(877833):
    if pd.isna(df["Activity"][i]):
        df.at[i,'Activity'] = "UNCONFIRMED"
```

```python
for i in range(877833):
    a = df['Changed'][i]
    b = df['Opened'][i]
    a = datetime.datetime.strptime(a,'%Y-%m-%d %H:%M:%S')  
    b = datetime.datetime.strptime(b,'%Y-%m-%d %H:%M:%S')  
    c = a - b
    c = c.total_seconds()/86400 # time in days.. I've kept it float continous values are better I guess (dunno?)
    df.at[i,'Total Time Open (in days)'] = c
```
